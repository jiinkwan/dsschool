## 행렬

### 열(Column) 가져오기

```r
# Survive 컬럼 가져오기
head(train["Survived"])
```

```r
# 여러개 컬럼 가져오기
head(train[c("Sex", "Name", "Survived")])
```

```r
# 리스트를 변수에 담아서 가져오기
columns <- c("Sex", "Name", "Survived")
head(train[columns])
```

### 행(row) 가져오기

```r
# 1번 행 가져오기
train[1,]
```

```r
# 1-7 번 행 가져오기
train[1:7,]
```

```r
# 1,3,7,13 행
train[c(1,3,7,13),]
```

```r
# 변수에 담아서 행 가져오기
passenger_ids <- c(1, 3, 7, 13)
train[passenger_ids,]
```

### 행렬 동시에 가져오기

```r
train[1, "Sex"]
```

```r
# 여러 열 가져오기
train[1, c("Sex", "Name", "Survived")]
```

```r
# 여러 행 가져오기
train[1:7, "Sex"]
```

```r
# 여러 행과 열을 동시에 가져오기
train[c(1,3,7,13), c("Sex","Name","Survived")]
```

## 색인

```r
# 남성 승객 조회
head(train[train["Sex"] == "male",])
```

```r
# 운임요금(Fare) 값이 20달러보다 높은
head(train[train["Fare"] > 20,])
```

```r
# 선착장(Embarked)이 C(Cherbourg)인 승객만 조회
head(train[train["Embarked"] == "C",])
```

```r
#head(train[train["Embarked"] %in% c("Q","S"),])
head(train[train["Embarked"] == "S",])
```

```r
head(train[train["Embarked"] == "S",])
```

```r
head(train[train["Embarked"] == c("Q","S"),])
```

```r
not_cherbourg <- c("Q", "S")
embarked_isin <- train["Embarked"] == not_cherbourg

head(train[embarked_isin,])
```
