```python
import pandas as pd
```

## Load Dataset

```python
train = pd.read_csv("data/titanic/train.csv")

print(train.shape)
train.head()
```

## Explore

```python
%matplotlib inline
import seaborn as sns
```

### countplot

```python
sns.countplot(data=train, x = "Embarked", hue = "Survived")
```

### barplot

```python
sns.countplot(data = train, x = "Pclass", hue = "Survived")
```

```python
sns.barplot(data = train, x = "Pclass", y = "Fare", hue = "Survived")
```

숫자에 대한 것은 count plot 대신 bar plot을 쓴다.

### pointplot

```python
sns.pointplot(data = train, x = "Pclass", y = "Fare", hue = "Survived")
```

Bar plot과 Point plot은 값 간의 연관성이 있으면, Point (시계열과 같은), 없다면 Bar plot(등급, titanic의 embarked)를 쓴다

### lmplot

```python
sns.lmplot(data = train, x = "Age", y = "Fare", hue = "Survived", fit_reg = False)
```

```python
## Outlier Removal
low_fare = train[train["Fare"] < 100]
sns.lmplot(data = low_fare, x = "Age", y = "Fare", hue = "Survived", fit_reg = False)
```

### distplot

```python
sns.distplot(low_fare["Fare"])
```
